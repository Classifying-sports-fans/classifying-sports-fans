```python
#Basic Packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Feed forward NN modules

import tensorflow as tf
import keras
from keras import layers
from tensorflow import convert_to_tensor
```

```python
## Importing training data, 555 from random state
## Train / test split done in another worksheet

fan_train = pd.read_csv('fan_train_555.csv', index_col=0)
```

```python
## rel for min set of 'relevant' features
rel_features = ['S2', 'D4', 'Fan_magnitude']

## VL1s of interest for prediction
VL1s = ['VL1r1','VL1r2','VL1r4','VL1r5',
       'VL1r7','VL1r10','VL1r11','VL1r12',
       'VL1r13','VL1r14']
```

```python
## Defining X and y

X = fan_train

#y = fan_train[VL1s]
```

```python
from sklearn.model_selection import train_test_split
```

```python
## Training and validation splits, conversion to tensors
X_tt, X_val = train_test_split(X, test_size=0.2, random_state=555, shuffle=True)

X_tt_tensor = convert_to_tensor(X_tt[rel_features])
X_val_tensor = convert_to_tensor(X_val[rel_features])
y_tt_tensor = convert_to_tensor(X_tt[VL1s])
y_val_tensor = convert_to_tensor(X_val[VL1s])
```

```python
# Compare this model with logistic regression because of sigmoid and other setup

model = keras.Sequential([
    #layers.Normalization(axis=-1), # Basically StandardScaler()
    #layers.Dense(20, activation = 'relu', use_bias = True), # R^2 --> R^12 with ReLU activation
    layers.Dense(50, activation = 'relu', use_bias = True), # R^12 --> R^12 with ReLU activation
    layers.Dense(1, use_bias = True, activation = 'sigmoid') # R^12 --> R^1 again with sigmoid activation
])

model.compile(optimizer = "rmsprop", 
              loss = "binary_crossentropy", 
              metrics = ["accuracy"])

for v in VL1s:
    model.fit(X_tt_tensor, X_tt[v], epochs = 50, validation_data=(X_val_tensor, X_val[v]))
    print(v, model.get_metrics_result())
```
